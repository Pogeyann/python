```python
import numpy as np 
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/DS/PYTHON/python_libraries/Pandas/files/customer',sep=',',header=None)
df.columns=['id','fname','lname','age','prof','location']
#find age mxm 2 employees fname lname age and location
df0=df.loc[(df['age'])]
print('*'*100)
df1=df.sort_values(by='age') [['fname','lname','age','location']]
print(df1.tail(2)) 
print('*'*100)
#find age min 1 employee
df2=df.sort_values(by='age') [['fname','lname','age','location']]
print(df2.head(1))
print('*'*100)
#india work and age above 40 fname,lname,age,prof,city
df3= df.loc[(df['location']=='india')&(df['age']>40)] [['fname','lname','age','prof','location']]
print(df3)
print('*'*100)
#us data collect
df4=df.loc[df['location']=='us']
print(df4)
print('*'*100)
#us age mxm 5 employee fname,lname,age,prof,city
df5=df.sort_values(by='age',ascending=False).head(5) [['fname','lname','age','prof','location']]
print(df5)
print('*'*100)
#india age minimum 3 employee details
df6=df.sort_values(by='age').head(3)
print(df6)
print('*'*100)
#age 40-60 between fname,lname,age,prof
df7 = df.loc[(df['age']>40)&(df['age']<60)] [['fname','lname','age','prof']]
print(df7)
#uk and age in between 25 to 35
df8 = df.loc[(df['location']=='uk')&(df['age']>25)&(df['age']<35)]
print(df8)
print('*'*100)
#australia age mxm 1 employee 
df9 = df.loc[df['location']=='australia'].sort_values(by='age',ascending=False).head(1)
print(df9)
print('*'*100)
#india and prof doctor age min 1 employee fname lname age
df10=df.loc[(df['location']=='india')&(df['prof']=='Doctor')].sort_values(by='age')[['fname','lname','age']].head(1)
print(df10)
```
